```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2],
                                                   df.iloc[:,-1],
                                                    test_size = 0.2)
```

```python
X_train.shape, X_test.shape,y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
Obj = OrdinalEncoder(categories = [['Average', 'Good', 'Best'], ['HS', 'UG', 'PG']])
```

```python
Obj.fit(X_train)
```

```python
X_train_encoded = Obj.transform(X_train)
X_test_encoded = Obj.transform(X_test)
```
